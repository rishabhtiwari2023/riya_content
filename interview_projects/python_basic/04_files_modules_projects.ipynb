```python
# Exception handling example
# Demonstrates try/except/finally which helps catch and handle errors gracefully
try:
    x = int('abc')  # this will raise ValueError
except ValueError as e:
    # Handle the specific ValueError and show a helpful message
    print('Caught ValueError:', e)
finally:
    # finally block runs whether or not an exception occurred
    print('This runs always (finally)')
```

```python
# Simple class example
# A class groups data (attributes) and behavior (methods) together
class Student:
    def __init__(self, name, math, science):
        # store the student's name and scores as attributes
        self.name = name
        self.math = math
        self.science = science
    def average(self):
        # compute and return the average of math and science
        return (self.math + self.science) / 2

# Create an instance and call the method to see how classes work
s = Student('Riya', 92, 88)
print(s.name, 'avg =', s.average())
```

```python
# Mini Project 1: Number Guessing Game (Starter)
# This demo shows the game logic. Replace the simulated guess with input() to play interactively.
import random

def play_guessing_game(max_tries=5):
    # pick a random secret number between 1 and 50
    secret = random.randint(1, 50)
    tries = 0
    while tries < max_tries:
        tries += 1
        # For interactive play replace the next line with: guess = int(input('Your guess: '))
        guess = random.randint(1, 50)  # simulated guess for demo
        print(f'Try {tries}: guessed {guess}')
        if guess == secret:
            print('Correct!')
            return True
        elif guess < secret:
            print('Too low')
        else:
            print('Too high')
    print('Out of tries. Secret was', secret)
    return False

# Demo run (non-interactive)
play_guessing_game(3)
```

```python
# Mini Project 2: Gradebook (Starter)
# Shows how to write and read CSV files and compute student averages
import csv

filename = 'gradebook.csv'
# Create a small sample gradebook (write CSV file)
with open(filename, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['name','math','science'])
    writer.writerow(['Riya', 92, 88])
    writer.writerow(['Anu', 85, 78])

# Read the CSV back and compute average scores per student
students = []
with open(filename, 'r', newline='') as f:
    reader = csv.DictReader(f)
    for row in reader:
        # Convert string fields to integers for math and science
        row['math'] = int(row['math'])
        row['science'] = int(row['science'])
        row['avg'] = (row['math'] + row['science'])/2
        students.append(row)

print('Students and averages:')
for s in students:
    print(s['name'], s['avg'])
```

```python
# Unit checks for notebook 4
# Simple test to ensure Student.average() returns the expected value
s = Student('Test', 80, 90)
assert s.average() == 85.0  # average should be (80+90)/2
print('OOP and project checks passed âœ…')
```

### Activity 2 â€” Math Quiz ðŸ§ âž•

Try it: run the quiz function below. For automatic runs we use `answers` list; to play interactively replace `answers` with `input()` prompts.


```python
# Try-it: Math Quiz (non-interactive demo)
# This small quiz shows questions and checks answers. Replace 'answers' with input() for real play.
questions = [
    ("2 + 3", 5),
    ("7 - 4", 3),
    ("3 * 3", 9)
]
# Simulated answers (replace with input() to prompt the user)
answers = [5, 3, 9]

score = 0
for (q, correct), ans in zip(questions, answers):
    print('Q:', q, 'Your answer:', ans)
    if ans == correct:
        score += 1
print('Score:', score, '/', len(questions))
```

```python
# Solution for Math Quiz: function and test
# run_quiz counts correct answers by comparing given answers to correct ones

def run_quiz(questions, answers):
    score = 0
    for (q, correct), ans in zip(questions, answers):
        if ans == correct:
            score += 1
    return score

# Test the function with simulated answers
assert run_quiz(questions, [5,3,9]) == 3
print('run_quiz test passed âœ…')
```

### Activity 3 â€” ASCII Dance ðŸ’ƒðŸ•º

Try it: this uses loops to draw a simple "dance" pattern. Change `steps` to make it longer or shorter.


```python
# Try-it: ASCII Dance (run this cell)
# This small loop prints an alternating ASCII art to look like a dancing figure.
import time

steps = 6  # number of dance steps; increase to make it longer
for i in range(steps):
    if i % 2 == 0:
        print("  o/   ")  # pose A
        print(" /|   ")
        print(" / \\  ")
    else:
        print("   o  ")  # pose B
        print("  \\|\\ ")
        print("  / \\ ")
    time.sleep(0.2)  # small pause to make the pattern readable

print('\nMake it longer by increasing `steps`!')
```
