```python
# Exercise 2: Swap two variables without third variable (Try it)
# Task: swap values of a and b. Try to write your swap before running the solution below.
a = 5
b = 9
print('Before:', a, b)
# (Your swap code goes here)
```

```python
# Solution to Exercise 2: Swap using tuple unpacking
# Explanation: Python can swap two variables in one line using multiple assignment.
a = 5
b = 9
print('Before:', a, b)
a, b = b, a  # swap values
print('After:', a, b)

# This is clearer and shorter than using a temporary third variable.
```

```python
# Exercise 3: Area of rectangle (Try it)
# Formula: area = length * width
length = 7  # change to test other values
width = 4
area = length * width
print('Area =', area)  # prints the computed area

# Solution note: multiply length and width to get the area of a rectangle.
```

```python
# Exercise 4: Celsius to Fahrenheit (Try it)
# Formula to remember: F = C * 9/5 + 32
C = 25  # change this to test other temperatures
F = C * 9 / 5 + 32
print(f"{C}Â°C = {F}Â°F")  # prints the converted temperature

# This shows how arithmetic and order of operations work in Python.
```

```python
# Exercise 5: Extract digits of a 3-digit number (Try it)
# Use floor division (//) and modulo (%) to get each digit
num = 123  # change to other 3-digit numbers
d1 = num // 100       # hundreds digit
d2 = (num // 10) % 10 # tens digit
d3 = num % 10         # units digit
print('Digits:', d1, d2, d3)

# This technique works for fixed-length numbers and helps understand integer division.
```

```python
# Exercise 6: Compute n! (iterative)
# Iterative approach multiplies numbers from 1 to n to get factorial
n = 5  # try other small values like 4 or 6
fact = 1
for i in range(1, n+1):
    fact *= i  # multiply fact by i each loop
print(f"{n}! = {fact}")

# Iterative solution avoids recursion for beginners and is easy to follow.
```

```python
# Unit tests (asserts) to check some functions
# These assertions check the sample outputs automatically â€” if an assert fails, fix the code above.
# Checking factorial for n=5
n = 5
fact = 1
for i in range(1, n+1):
    fact *= i
assert fact == 120, 'Factorial for 5 should be 120'

# Check swapping sample using tuple unpacking
a, b = 5, 9
a, b = b, a
assert (a, b) == (9, 5)

print('Basic tests passed âœ…')
```
